##### Load Modules

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
```

##### Extracting text from web page

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

##### Extracting text from html tags

```python
text = soup.get_text()
text
```

##### Cleaning data ; removing \n & \t

```python
text=text.replace('\n',' ')
text=text.replace('\t',' ')
text
```

##### Using nltk word tokenizer for geting words from text

```python
words = nltk.word_tokenize(text)
words
```

##### Using nltk FreqDist function to get frequency of words

```python
counts = nltk.FreqDist(w.lower() for w in words) ## converting all words into lowercase
```

##### Total unique words

```python
counts.B()
```

##### frequency of words

```python
counts.most_common(376) ## get all words
```

##### plot for top 20

```python
counts.plot(20,cumulative=False)
```

##### word with hight frequency

```python
counts.max()
```
